```python
spotify.dropna(inplace=True)
spotify.duplicated().sum()
```

```python
features= ['track_id', 'artists', 'popularity', 'danceability', 'energy','loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'track_genre']
```

```python
spotify['track_genre'].value_counts()
```

```python
# Change the the Artist Column in a List

spotify['artists'] = spotify['artists'].apply(lambda x: x.split(','))


```

```python
# Replce ' ' with ''

spotify['artists'] = spotify['artists'].apply(lambda x: [i.replace(' ','') for i in x])
spotify.head()
```

```python
# If I see a ; in Artist value, I split it

spotify['artists'] = spotify['artists'].apply(lambda x: [i.split(';') for i in x])
```

```python
# Flatten the list show it in a single list

spotify['artists'] = spotify['artists'].apply(lambda x: [item for sublist in x for item in sublist])
spotify.head()
```

```python
new_df = spotify[features]
new_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import CountVectorizer

```

```python
# Convert the track_genre column to a list 
new_df['track_genre'] = new_df['track_genre'].apply(lambda x: x.split(','))
new_df.head()
```

```python
# Create new column called tag and join all the values of  artist and track_genre
new_df['tag'] = new_df['artists'] + new_df['track_genre']
new_df.head()
```

```python
# Replace a Commans with a space
new_df['tag'] = new_df['tag'].apply(lambda x: [i.replace(',',' ') for i in x])
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x: ' '.join(x))
new_df.head()
```

```python
# Import train test split
from sklearn.model_selection import train_test_split

# Split the data into train and test
train, test = train_test_split(new_df, test_size=0.8)

train.shape
```

```python
vectorizer = CountVectorizer(max_features=5000, stop_words='english')
tfidf_matrix = vectorizer.fit_transform(train['tag']).toarray() 
tfidf_matrix.shape
```

```python
scaler = MinMaxScaler()
X_numeric = scaler.fit_transform(train[['popularity', 'danceability', 'energy','loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']])
```

```python
X = np.concatenate((tfidf_matrix, X_numeric), axis=1)
cosine_similarities = cosine_similarity(X)
```

```python
neigh = NearestNeighbors(n_neighbors=3)
neigh.fit(cosine_similarities)
```

```python
def recommend_song(song_id):
    try:
        song_index = new_df[new_df['track_id'] == song_id].index[0]
        distances, indices = neigh.kneighbors(cosine_similarities[song_index].reshape(1, -1), n_neighbors=3)
        track_ids = []
        for i in indices[0][1:]:
            track_ids.append(new_df.iloc[i]['track_id'])
        return track_ids
    except:
        # Return Data by Popularity
        audio_features = ['danceability', 'energy','loudness','speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'popularity']
        feature_df = new_df[audio_features]
        distances, indices = neigh.kneighbors(feature_df, n_neighbors=3)
        track_ids = []
        for i in indices[0][1:]:
            track_ids.append(new_df.iloc[i]['track_id'])
        print("I am here")
        return track_ids


```

```python
print(recommend_song('45oxtVfleBBYG9ps80hyTL'))
```
